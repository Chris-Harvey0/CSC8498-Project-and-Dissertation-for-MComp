```python
from google.colab import drive
drive.mount('/content/drive')

import pandas as pd
import json

path = "/content/drive/MyDrive/CSC8498_Dissertation/Raw_Data/"
```

```python
years = ["2018", "2019", "2020", "2021", "2022"]
df = pd.DataFrame(columns=['date_time', 'intensity_forecast', 'intensity_actual'])

for year in years:
  print("Processing data for {}".format(year))
  month = 1
  while month < 13:
    f = open(path + year + "/" + str(month) + ".json", "rb")
    json_data = json.load(f)
    f.close()
    for row in json_data:
      midpoint_date_time = row['from'][:13] + str((int(row['from'][14:16]) + 15 ) % 60) # Get midpoint of time between 'from' and 'to'
      df.loc[len(df.index)] = [midpoint_date_time, row['intensity']['forecast'], row['intensity']['actual']]
    print("{:.2f}%".format((100/12)*month))
    month += 1
```

```python
df.to_csv(path + "2018-2022.csv")
```
